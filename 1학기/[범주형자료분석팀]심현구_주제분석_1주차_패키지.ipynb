```python
import requests

html = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%84%B1%EA%B7%A0%EA%B4%80%EB%8C%80'

response = requests.get(html)
```

```python
#문제 2. 위 코드를 참고하여 특정 키워드와 관련된 기사를 원하는 페이지까지 크롤링하여 페이지수,
#기사의 제목, 작성시점, 언론사, url 정보를 가진 데이터를 만들어봅시다.
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
page = []
date = []
press = []
title = []
url = []

for i in range(3) :
    url1 = f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%84%B1%EA%B7%A0%EA%B4%80%EB%8C%80&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=76&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={i*10+1}'
    
    response2 = requests.get(url1)
    soup2 = bs(response2.text , 'html.parser')
    
    page_html = soup2.select('div.sc_page_inner > a.btn')  
    date_html = soup2.select('span.info')
    press_html = soup2.select('a.info.press')
    title_html = soup2.select('a.news_tit')
    url_html = soup2.select('a.news_tit')

    for j in date_html :
        date.append(j.text)
    for j in press_html :
        press.append(j.text)
    for j in title_html :
        title.append(j.text)
        page.append(page_html[i].text)
    for j in url_html :
         url.append(j.attrs['href'])


list_sum = list(zip(page, date, press, title, url))
df = pd.DataFrame(list_sum, columns=['page', 'date', 'press', 'title', 'url'])
```

```python
import time

scroll_location = driver.execute_script("return document.body.scrollHeight")

while True:
#현재 스크롤의 가장 아래로 내림
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
       
    #전체 스크롤이 늘어날 때까지 대기
    time.sleep(2)
    
    #이동 후 스크롤 위치
    scroll_height = driver.execute_script("return document.body.scrollHeight")

    #늘어난 스크롤 위치와 이동 전 위치 같으면 종료
    if scroll_location == scroll_height:
        break   
    #같지 않으면 스크롤 위치 값을 수정하여 같아질 때까지 반복
    else:
        scroll_location = driver.execute_script("return document.body.scrollHeight")
```

```python
items = driver.find_elements(By.CSS_SELECTOR,".basicList_info_area__TWvzp")
list_name = []
list_price = []
list_url = []
list_ads = []
```

```python
import pandas as pd
scroll_location = driver.execute_script("return document.body.scrollHeight")

while True:
    try :
        for item in items:    
            name=item.find_element(By.CSS_SELECTOR,'.basicList_title__VfX3c').text
            try :
                price = item.find_element(By.CSS_SELECTOR,'.price_num__S2p_v').text
            except:
                price='판매중단'
            if(item.find_element(By.CSS_SELECTOR,'.basicList_title__VfX3c > a').get_attribute('data-testid')[-2:] == "AD") :
                list_ads.append('o')
            else :
                list_ads.append('x')
               
    
            link = item.find_element(By.CSS_SELECTOR,'.basicList_title__VfX3c > a').get_attribute('href')
    
            list_name.append(name)
            list_price.append(price)
            list_url.append(link)

        #현재 스크롤 내림
        dr.execute_script("window.scrollTo(0,{})".format(scroll_location + 648))
    
        #전체 스크롤이 늘어날 때까지 대기
        time.sleep(5)
    
        #이동 후 스크롤 위치
        scroll_height = driver.execute_script("return document.body.scrollHeight")

    #늘어난 스크롤 위치와 이동 전 위치 같으면 종료
        if scroll_location == scroll_height:
            break   
    #같지 않으면 스크롤 위치 값을 수정하여 같아질 때까지 반복
        else:
            scroll_location = driver.execute_script("return document.body.scrollHeight")
    except:
        break
        
df=pd.DataFrame({'제품명': list_name, '가격':list_price,'url':list_url, 'ads': list_ads})
df
```

```python
list_name = []
list_deptime = []
list_flytime = []
list_price = []
items = []

def flatten(list):
    result = []
    for j in list:
        result.extend(j)
    return result
```
